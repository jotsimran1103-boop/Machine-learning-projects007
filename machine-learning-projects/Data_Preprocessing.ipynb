```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
# add a target coulumn
df['target'] = data.target
df.head()
```

```python
df.isnull().sum()
```

```python
#Add  null values
df.loc[5:15, 'mean radius'] = None
df.loc[5:15, 'mean perimeter'] = None
```

```python
#now handle null values
from sklearn.impute import SimpleImputer
SI = SimpleImputer(strategy='mean')
df[['mean radius', 'mean perimeter']] = SI.fit_transform(df[['mean radius','mean perimeter']])
```

```python
# Check outliers and handle them
q1 = df['mean radius'].quantile(0.25)
q3 = df['mean radius'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
```

```python
outliers = df[(df['mean radius'] < lower_bound) | (df['mean radius'] > upper_bound)]
outliers.shape
```

```python
# Eliminate the outliers
df = df[(df['mean radius'] < lower_bound) & (df['mean radius'] > upper_bound)]
```

```python
## Enconding
df['tumor_size'] = pd.cut(
   df['mean radius'],
   bins = [0,12,18,30],
   labels = ['small','medium','large']
)

df['hospital_type'] = ['private' if i%2 == 0 else 'govt' for i in range (len(df))]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['tumor_size'] = le.fit_transform(df['tumor_size'])
df['hospital_type'] = le.fit_transform(df['hospital_type'])
```
